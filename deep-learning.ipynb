```python
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
import pandas as pd
```

# Seeding to make sure that we get exact same result everytime the code is run

```python
seed = 7
np.random.seed(seed)
```

# Predictive Analysis problem. Predict the chance that a person will buy a bike

```python
input_file = 'KerasDownloads/BBCN.csv'
input_df = pd.read_csv(input_file)
```

# Preview

```python
print input_df.shape
input_df.head()
```

# Split the data to input and target variable

```python
array = input_df.values
X = array[:, 0:11]
Y = array[:, 11]
```

# Build the model

Models in keras are defined as the sequence of layers

  First layer is the input layer and the subsequential are hidden layers except the output layer. This forms a `Sequential` model.
  
  Neurons in the layer are fully connected, meaning every neuron in ith layer is connected to every neuron in (i+1)th layer. For this we use `Dense` layering structure
  
  `input_dim` is the number of features used to predict the output
  
  

```python
model = Sequential()
model.add(Dense(11, input_dim = 11, kernel_initializer = 'uniform', activation = 'relu')) # input layer
model.add(Dense(8, kernel_initializer = 'uniform', activation = 'relu')) # 1st hidden layer
model.add(Dense(8, kernel_initializer = 'uniform', activation = 'relu')) # 2nd hidden layer
model.add(Dense(8, kernel_initializer = 'uniform', activation = 'relu')) # 3nd hidden layer
model.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid')) # output layer
```

 # Model compilation
 
 This uses the backend Tensorflow here.
 
 `metrics` are used to judge the performance of our model. Here we are using `accuracy` which is good only for symmetric data sets, where the class distribution is 50/50.In our case, classes are balanced.
 
 Use `crossentropy` cost function in case of classification problems.

```python
model.compile(loss = 'binary_crossentropy',optimizer = 'adam', metrics = ['accuracy'])
```

# Model fitting

`epochs` = number of times to run the model over the training data

`batch_size` = number of samples that going to be propagated through the network

See this: https://stats.stackexchange.com/questions/153531/what-is-batch-size-in-neural-network


```python
model.fit(X, Y, epochs = 200, batch_size = 10)
```

# Model score

```python
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
